```python
#question_no_2

def find_max_Length_of_CN(numbers):
    if not numbers:
        return 0
    
    #initlization varibale by 1
    max_length = 1
    current_length = 1

    for i in range(1, len(numbers)):
        if numbers[i] > numbers[i - 1]:  #checking if input number is continuous increasing or not
            current_length += 1
            max_length = max(max_length, current_length) 
        else:
            current_length = 1

    return max_length

number = [1,3,5,4,7]
find_max_Length_of_CN(number)
```

```python
#question_no_3

class Solution:
    def largestNumber(self, nums):
        
        #1st we are converting it in string
        nums = [str(num) for num in nums]
        
        
        def sort_key(num):
            repeated_num = num * 9
            return repeated_num, num
        
        #sort num based on sort_key
        sorted_nums = sorted(nums, key=sort_key, reverse=True)
        
        # Join the sorted strings, and return "0" if the result is "0"
        result = "".join(sorted_nums)
        return "0" if result[0] == "0" else result

Input_num =  [3,30,34,5,9]    
S = Solution()  #making object of class
S.largestNumber(Input_num)
```

```python
#question_no_1

#importing zipfile lib
import zipfile

#function for reading csv_file_inside_zip_file

def read_csv_file_inside_zip(zip_file_path,csv_file_name):
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        with zip_ref.open(csv_file_name) as csv_file:
            df = pd.read_csv(csv_file)
    return df        
```
